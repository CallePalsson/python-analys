```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
tabell = pd.read_csv("shl-dataset.csv",sep=",")
pd.set_option('display.max_rows', None)
print(tabell)

```

```python
plt.scatter(
  tabell["Rank"],
  tabell["EstimatedPlayerBudgetMSEK"],
)

plt.xlabel("Rank")
plt.ylabel("PlayerBudget")

plt.gca().invert_xaxis()

plt.plot()


```

```python
plt.scatter(
    tabell["EstimatedPlayerBudgetMSEK"],
    tabell["SeasonPoints"]
)

plt.xlabel("Player Budget (MSEK)")
plt.ylabel("Season Points")

plt.title("SHL 25/26: Player Budget vs Season Points")

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.stem(
    tabell["Team"],
    tabell["SeasonPoints"]
)

plt.xticks(rotation=45)
plt.xlabel("Team")
plt.ylabel("Season Points")
plt.title("SHL 25/26 Season Points by Team")

plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = tabell["EstimatedPlayerBudgetMSEK"]
y = tabell["SeasonPoints"]

# regressionslinje
m, b = np.polyfit(x, y, 1)

plt.figure(figsize=(10,6))

# scatterplot
plt.scatter(x, y)

# trendlinje
plt.plot(x, m*x + b)

# lag-namn
for i, row in tabell.iterrows():
    plt.text(
        row["EstimatedPlayerBudgetMSEK"],
        row["SeasonPoints"],
        row["Team"],
        fontsize=8
    )

plt.xlabel("Player Budget (MSEK)")
plt.ylabel("Season Points")
plt.title("SHL 25/26 Budget vs Performance")

plt.grid()

plt.show()
```
